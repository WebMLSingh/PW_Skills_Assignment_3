1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

numbers = []
n = int(input("Enter the number of elements in the list: "))
for i in range(0,n):
    ele=int(input("Enter the elements in the List: "))
    numbers.append(ele)
result = sum_of_even_numbers(numbers)
print(result)
```

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
def square_numbers(numbers):
    return [num ** 2 for num in numbers]

numbers = []
n = int(input("Enter the number of elements in the list: "))
for i in range(0,n):
    ele=int(input("Enter the elements in the List: "))
    numbers.append(ele)
result = square_numbers(numbers)
print(result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

number = int(input("Enter a number: "))
result = is_prime(number)
if result:
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.current_term = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.n_terms:
            raise StopIteration
        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:
            self.current_term += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b

n = int(input("Enter the number of terms: "))
fib_iterator = FibonacciIterator(n)
for fib in fib_iterator:
    print(fib)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

for power in powers_of_two(5):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
      for line in file:
        yield line

line_in_file = read_file_line_by_line('Test.txt')
# for line in line_in_file:
#     print(line)
print(next(line_in_file))
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
# Example list of tuples
tuples_list = [(1, 3), (4, 1), (2, 5), (6, 2)]

# Sorting using lambda function based on the second element
sorted_list = sorted(tuples_list, key=lambda x: x[1])

# Example usage
print(sorted_list)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Using map() to convert each Celsius temperature to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Taking Input as a string
input_string = input("Enter a string: ")

# Using filter() to remove vowels
filtered_string = ''.join(filter(is_not_vowel, input_string))

print(filtered_string)
```
